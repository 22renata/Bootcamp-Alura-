Desafio: escolher um título mais **descritivo**, que passe a mensagem adequado.

```
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x="Unidade da Federação", y="2020/Ago", kind="bar", figsize=(9,6), color="r")
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valor gasto com atendimento hospitalar por unidade da federação")
plt.show()
```
